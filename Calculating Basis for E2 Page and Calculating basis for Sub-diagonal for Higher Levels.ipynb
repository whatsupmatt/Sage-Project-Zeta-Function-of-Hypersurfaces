```python
R.<w,x,y,z>=QQ[]
f(w,x,y,z)=#insert equation of hypersurface
f1(w,x,y,z)=derivative(f,w)
f2(w,x,y,z)=derivative(f,x)
f3(w,x,y,z)=derivative(f,y)
f4(w,x,y,z)=derivative(f,z)
#we define multiplication by the w derivative of f, x derivative of f, ....
def Multiplication1(list_in1):
    ImPartialw=[]
    for poly in list_in1:
        Result1= expand(poly*f1(w,x,y,z))
        ImPartialw.append(Result1)
    return ImPartialw


def Multiplication2(list_in2):
    ImPartialx=[]
    for poly in list_in2:
        Result2=expand(poly*f2(w,x,y,z))
        ImPartialx.append(Result2)
    return ImPartialx


def Multiplication3(list_in3):
    ImPartialy=[]
    for poly in list_in3:
        Result3=expand(poly*f3(w,x,y,z))
        ImPartialy.append(Result3)
    return ImPartialy


def Multiplication4(list_in4):
    ImPartialz=[]
    for poly in list_in4:
        Result4=expand(poly*f4(w,x,y,z))
        ImPartialz.append(Result4)
    return ImPartialz

#We define a function that allows us to append rows to a matrix
def insert_row(Matrix,k,row):
    return matrix(Matrix.rows()[:k]+[row]+Matrix.rows()[k:])

#We define the matrix representing df wedge from 3 forms to 4 forms (m=0, n=2)
def A(h,m,n):
    k1=binomial(4+m-1,m)
    k2=binomial(4+n-1,n)
    degree1 = m; [R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree1, length=4)]

    degree2= n; [R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree2, length=4)]
    List1=[R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree1, length=4)]
    List2=[R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree2, length=4)]
    
    CoefficientwList=[]
    for poly in Multiplication1(List1):
        for mono in List2:
            Resultw=poly.coefficient(mono)
            CoefficientwList.append(Resultw)


    CoefficientxList=[]
    for poly in Multiplication2(List1):
        for mono in List2:
            Resultx=poly.coefficient(mono)
            CoefficientxList.append(Resultx)


    CoefficientyList=[]
    for poly in Multiplication3(List1):
        for mono in List2:
            Resulty=poly.coefficient(mono)
            CoefficientyList.append(Resulty)


    CoefficientzList=[]
    for poly in Multiplication4(List1):
        for mono in List2:
            Resultz=poly.coefficient(mono)
            CoefficientzList.append(Resultz)

    def Separation1(list_in11):
        DivCoefficientfwList=[]
        x=1
        while x<k1+1:
            DivCoefficientfwList.append(list_in11[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfwList

    def Separation2(list_in22):
        DivCoefficientfxList=[]
        x=1
        while x<k1+1:
            DivCoefficientfxList.append(list_in22[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfxList

    def Separation3(list_in33):
        DivCoefficientfyList=[]
        x=1
        while x<k1+1:
            DivCoefficientfyList.append(list_in33[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfyList

    def Separation4(list_in44):
        DivCoefficientfzList=[]
        x=1
        while x<k1+1:
            DivCoefficientfzList.append(list_in44[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfzList

    Separation1NfwList=[]
    j1=0
    while j1<k1:
        ResultN1=[-1*P for P in Separation1(CoefficientwList)[j1]]
        j1 +=1
        Separation1NfwList.append(ResultN1)


    Separation2NfxList=[]
    j2=0
    while j2<k1:
        ResultN2=[-1*P for P in Separation2(CoefficientxList)[j2]]
        j2 +=1
        Separation2NfxList.append(ResultN2)


    Separation3NfyList=[]
    j3=0
    while j3<k1:
        ResultN3=[-1*P for P in Separation3(CoefficientyList)[j3]]
        j3 +=1
        Separation3NfyList.append(ResultN3)


    Separation4NfzList=[]
    j4=0
    while j4<k1:
        ResultN4=[-1*P for P in Separation4(CoefficientzList)[j4]]
        j4 +=1
        Separation4NfzList.append(ResultN4)


    #ordering in domain goes all monomials dwdxdy, "" dwdxdz, "" dwdydz, ""dxdydz
    #ordering in range is clear, all monomials dwdxdydz

    M=Matrix(QQ, 4*k1,k2)
    i1=0
    while i1<k1:
        M[i1,0:k2]=vector(Separation4(CoefficientzList)[i1])
        i1 +=1

    i2=0
    while i2<k1:
        M[i2+k1,0:k2]=vector(Separation3NfyList[i2])
        i2 +=1

    i3=0
    while i3<k1:
        M[i3+2*k1,0:k2]=vector(Separation2(CoefficientxList)[i3])
        i3 +=1

    i4=0
    while i4<k1:
        M[i4+3*k1,0:k2]=vector(Separation1NfwList[i4])
        i4 +=1
    return M
#we define the matrix for df wedge from 2 forms to 3 forms
def B(h,m,n):
    k1=binomial(4+m-1,m)
    k2=binomial(4+n-1,n)
    degree1 = m; [R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree1, length=4)]

    degree2= n; [R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree2, length=4)]
    List1=[R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree1, length=4)]
    List2=[R.monomial(*[x-1 for x in c]) for c in Compositions(4+degree2, length=4)]
    



    CoefficientwList=[]
    for poly in Multiplication1(List1):
        for mono in List2:
            Resultw=poly.coefficient(mono)
            CoefficientwList.append(Resultw)


    CoefficientxList=[]
    for poly in Multiplication2(List1):
        for mono in List2:
            Resultx=poly.coefficient(mono)
            CoefficientxList.append(Resultx)


    CoefficientyList=[]
    for poly in Multiplication3(List1):
        for mono in List2:
            Resulty=poly.coefficient(mono)
            CoefficientyList.append(Resulty)


    CoefficientzList=[]
    for poly in Multiplication4(List1):
        for mono in List2:
            Resultz=poly.coefficient(mono)
            CoefficientzList.append(Resultz)

    def Separation1(list_in11):
        DivCoefficientfwList=[]
        x=1
        while x<k1+1:
            DivCoefficientfwList.append(list_in11[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfwList

    def Separation2(list_in22):
        DivCoefficientfxList=[]
        x=1
        while x<k1+1:
            DivCoefficientfxList.append(list_in22[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfxList

    def Separation3(list_in33):
        DivCoefficientfyList=[]
        x=1
        while x<k1+1:
            DivCoefficientfyList.append(list_in33[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfyList

    def Separation4(list_in44):
        DivCoefficientfzList=[]
        x=1
        while x<k1+1:
            DivCoefficientfzList.append(list_in44[(x-1)*k2:k2*x])
            x += 1
        return DivCoefficientfzList

    Separation1NfwList=[]
    j1=0
    while j1<k1:
        ResultN1=[-1*P for P in Separation1(CoefficientwList)[j1]]
        j1 +=1
        Separation1NfwList.append(ResultN1)


    Separation2NfxList=[]
    j2=0
    while j2<k1:
        ResultN2=[-1*P for P in Separation2(CoefficientxList)[j2]]
        j2 +=1
        Separation2NfxList.append(ResultN2)


    Separation3NfyList=[]
    j3=0
    while j3<k1:
        ResultN3=[-1*P for P in Separation3(CoefficientyList)[j3]]
        j3 +=1
        Separation3NfyList.append(ResultN3)


    Separation4NfzList=[]
    j4=0
    while j4<k1:
        ResultN4=[-1*P for P in Separation4(CoefficientzList)[j4]]
        j4 +=1
        Separation4NfzList.append(ResultN4)
        
    M2=Matrix(QQ, 6*k1,4*k2)
    i1=0
    while i1<k1:
        M2[i1,0:k2]=vector(Separation3(CoefficientyList)[i1])
        M2[i1,k2:2*k2]=vector(Separation4(CoefficientzList)[i1])
        i1 +=1

    i2=0
    while i2<k1:
        M2[i2+k1,0:k2]=vector(Separation2NfxList[i2])
        M2[i2+k1,2*k2:3*k2]=vector(Separation4(CoefficientzList)[i2])
        i2 +=1

    i3=0
    while i3<k1:
        M2[i3+2*k1,k2:2*k2]=vector(Separation2NfxList[i3])
        M2[i3+2*k1,2*k2:3*k2]=vector(Separation3NfyList[i3])
        i3 +=1

    i4=0
    while i4<k1:
        M2[i4+3*k1,0:k2]=vector(Separation1(CoefficientwList)[i4])
        M2[i4+3*k1,3*k2:4*k2]=vector(Separation4(CoefficientzList)[i4])
        i4 +=1

    i5=0
    while i5<k1:
        M2[i5+4*k1,k2:2*k2]=vector(Separation1(CoefficientwList)[i5])
        M2[i5+4*k1,3*k2:4*k2]=vector(Separation3NfyList[i5])
        i5 +=1

    i6=0
    while i6<k1:
        M2[i6+5*k1,2*k2:3*k2]=vector(Separation1(CoefficientwList)[i6])
        M2[i6+5*k1,3*k2:4*k2]=vector(Separation2(CoefficientxList)[i6])
        i6 +=1
    return M2

```
